```python
# Import the functions for downloading data from NWIS
import dataretrieval.nwis as nwis

# Specify the USGS site code
site = '03430320' # CUMBERLAND RIVER AT BRILEY PKWY NR INGLEWOOD, TN

# Get instantaneous values (iv)
df = nwis.get_record(sites=site, service='dv', start='2014-06-05', end='2015-06-04')
df
```
